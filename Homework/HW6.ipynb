1.	A whiffle ball is dropped from the UD Tower, a height of 57 m above the landscaped ground.  Using  Heun’s method  and assuming a terminal velocity of 10 m/s determine 
a.	the speed of the whiffle ball at any instant while it is falling,  
b.	the time to reach the ground,
c.	the time to reach the ground in the absence of air resistance.


```python
"""
Question 1
"""
import numpy as np
import matplotlib.pyplot as plt


grav = 9.81
initVel = 0
termVel = 10
time = 0
height = 57
vel = initVel
N = 20
dt = 0.001
timeVel = {}
heightArr = np.zeros(0)
timeArr = np.zeros(0)
"""
dv / dt
"""


def acc(t):
    return -grav * (1 + (vel * np.abs(vel)) / termVel**2)


def accNoTV(v):
    return v - grav


"""
Heun's Method
"""


def Heun_V(v):
    heunV = v + acc(v) * dt
    v = v + (acc(v) + acc(heunV)) * dt / 2
    return v


def Heun_NoTV(v):
    heunV = v + accNoTV(v) * dt
    v = v + (accNoTV(v) + accNoTV(heunV)) * dt / 2
    return v


def Heun_Y(y, v):
    heunY = y + v * dt
    y = y + (v * v) * dt/2
    return heunY


def plotWithTV():
    global vel, height, heightArr, time, timeArr
    while height >= 0:
        vel = Heun_V(vel)
        height = Heun_Y(height, vel)
        heightArr = np.append(heightArr, height)
        time = time + dt
        timeVel[time] = vel
        timeArr = np.append(timeArr, time)
    print(
        "The time it takes to reach the ground is with air resistane: ", timeArr[-2])
    plt.plot(timeArr, heightArr, "g-")
    plt.show()

# plotWithTV()


def plotNoTV():
    global vel, height, heightArr, time, timeArr
    while height >= 0:
        vel = Heun_NoTV(vel)
        height = Heun_Y(height, vel)
        heightArr = np.append(heightArr, height)
        time = time + dt
        timeVel[time] = vel
        timeArr = np.append(timeArr, time)
    print(
        "The time it takes to reach the ground is without air resistance is: ", timeArr[-2])
    plt.plot(timeArr, heightArr, "g-")
    plt.show()
# plotNoTV()


"""
a. finds speed of ball at any instant while falling
"""


def find_closest_number(arr, target):
    closest = None
    min_difference = float('inf')

    for num in arr:
        difference = abs(num - target)
        if difference < min_difference:
            min_difference = difference
            closest = num

    return closest


def findSpeed(t):
    if t in timeVel:
        print("The speed of the ball at", t,
              " seconds is: ", timeVel[t], "m/s")
    else:
        temp = find_closest_number(timeArr, t)
        print("The speed of the ball at", t,
              "seconds is roughly ", timeVel[temp], "m/s")


findSpeed(9)
```
